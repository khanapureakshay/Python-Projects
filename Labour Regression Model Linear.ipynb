## Optional Step: Checking for P-value Using  StatsModel

```python
import statsmodels.api as sm #This stands for Stats Model
X_train_sm= X_train
# Unlikely of SKlearn the Statsmodles do not fit a constant so we need to add Constant
X_train_sm=sm.add_constant(X_train_sm)
# Create a fitted Model in one Line
lm_1=sm.OLS(y_train,X_train_sm).fit()
# Print Coefficient
lm_1.params
```

1. F- statistics : Are the OLS models built significantly differnt from each other (Higher the better)
2. Prob(F- stats): P-value for the F-stats is define wheather F-stats Significant or not
3. Log-Likelihood : Is coverage close, closer the value to 0 indicates that Convergence point is nearer
4. AIC,BIC: Model goodnesss of fit vs Complexity
         If the value is low indicates Model is Close to Best Fit (True  Model)
5. Skewness: Measure of symm of residual around mean 
6. Kurtosis : Measure of shape of distribution of residual compare residuals that are closer vs residual that are far away from mean 
7. Omnibus: Combine Kurtosis and Skewness and provide on estimate which is statistical in nature
8. Jorque bera : Different tst test for skewness and kurtosis of REsidual
9. Durbin-Watson: A test for the presence of auto correlation of residual and IMP in Time Series
